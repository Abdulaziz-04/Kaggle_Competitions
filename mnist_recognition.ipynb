```python
X=pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
xtest=pd.read_csv('/kaggle/input/digit-recognizer/test.csv')
sample=pd.read_csv('/kaggle/input/digit-recognizer/sample_submission.csv')
```

```python
#Output labels for train dataset
y=X['label']
#Input features for the train dataset
xtrain=X.drop(['label'],axis=1)
xtrain
```

```python
y=to_categorical(y)
y.shape
```

```python
xtrain=xtrain/255
xtest=xtest/255
```

```python
arr_xtrain=xtrain.loc[:,:].values
arr_xtrain=arr_xtrain.reshape(arr_xtrain.shape[0],28,28,1)
arr_xtrain.shape
```

```python
arr_xtest=xtest.loc[:,:].values
arr_xtest=arr_xtest.reshape(arr_xtest.shape[0],28,28,1)
arr_xtest.shape
```

```python
model=Sequential()
model.add(Conv2D(filters=32,kernel_size=(4,4),input_shape=(28,28,1),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='rmsprop',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(arr_xtrain,y,epochs=15)
```

```python
results=np.argmax(model.predict(arr_xtest),axis=-1)
results
```

```python
final_df=pd.DataFrame({'ImageId':sample['ImageId'],'Label':results})
final_df.to_csv('submission.csv',index=False)
```
